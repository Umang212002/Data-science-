```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.amazon.in/Apple-iPhone-128GB-Space-Black/product-reviews/B0BDJ22G36/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
soup
```

```python
names=soup.find_all('span',class_='a-profile-name')
names
```

```python
len(names)
```

```python
import pandas as pd
df=pd.DataFrame(names,columns={'Customer names'})
df
```

```python
df=df.drop_duplicates(ignore_index=True)
df
```

```python
df=df.drop(2,axis=0)
df
```

```python
title=soup.find_all('a',class_='review-title')[1:-1]
title
```

```python
len(title)
```

```python
#Data filtering 
#Data cleaning
r_title=[]
for i in range(0,len(title)):
  r_title.append(title[i].get_text()[1:-1])
r_title  
```

```python
len(r_title)
```

```python
date=soup.find_all('span',class_='review-date')[2:]
date
```

```python
review_date=[]
for i in range(0,len(date)):
  #review_date.append(date[i].get_text()) # for full data 
  review_date.append(date[i].get_text().replace('Reviewed in India üáÆüá≥ on',''))

review_date
```

```python
df
```

```python
len(r_title)
```

```python
df['Review Title']=r_title
df['Review Dates']=review_date
df
```

```python
rating=soup.find_all('i',class_='review-rating')[2:]
rating
```

```python
len(rating)
```

```python
r_rating=[]
for i in range(0,len(rating)):
  r_rating.append(rating[i].get_text())
r_rating
```

```python
df
```

```python
df['Rating']=r_rating
df
```
